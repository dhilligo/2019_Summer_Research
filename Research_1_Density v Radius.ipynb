```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
X=np.asarray(np.loadtxt('/Users/dylanhilligoss/vh1/tdsw1.init', skiprows=1)).T
```

```python
Radius=X[2]
Density=X[3]

#print(len(Radius),len(Density))
```

```python
#plt.plot(x,y,**kwargs)

fig=plt.figure(figsize=(5,5))
plt.rcParams.update({'font.size':14, 'font.family':'Times new roman'})

plt.plot(Radius,Density, color='b', ls='-')

plt.title("Density vs. Radius")
plt.ylabel("Density")
plt.xlabel("Radius")

plt.tight_layout()
```

```python
#print(X)
#plt.plot(x,y,**kwargs)

fig=plt.figure(figsize=(9,4))
plt.rcParams.update({'font.size':14, 'font.family':'Times new roman'})

ax0 = fig.add_subplot(1,2,1)
ax0.plot(Radius,Density, color='b', ls='-')
ax0.set_title("Density vs. Radius")
ax0.set_ylabel("Density")
ax0.set_xlabel("Radius")

ax1 = fig.add_subplot(1,2,2)
ax1.plot(Radius,X[4], color='y', ls=':')
ax1.set_title("Temperature? vs. Radius")
ax1.set_ylabel("Temperature")
ax1.set_xlabel("Radius")


plt.tight_layout(w_pad=1.5)
```
